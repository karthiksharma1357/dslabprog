```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class Stack:
  def __init__(self,l):
    self.top=None
    self.l=l
  def push(self,data):
     Nn=Node(data)
     d=0
     Cn=self.top
     while(Cn!=None):
      d+=1
      Cn=Cn.next
     if self.top==None:
      self.top=Nn
      self.top.next=None
     elif d>=self.l:
      print("Stack Overflow")
     else:
      Nn.next=self.top
      self.top=Nn
  def pop(self):
    if self.top==None:
      print("No items to delete")
    elif self.top.next==None:
      print(self.top.data)
      self.top=None
    else:
      temp=self.top
      print(self.top.data)
      self.top=temp.next
      temp=None
  def disp(self):
    Cn=self.top
    if self.top==None:
      print("No data in stack")
    else:
      while(Cn!=None):
        print(Cn.data)
        Cn=Cn.next
  def is_empty(self):
    if self.top==None:
      print("No data in stack")
    else:
      print(self.top.data)
z=int(input("Enter the number of elemnts in stack"))
S1=Stack(z)
while(True):
  x=int(input("1:push 2:pop 3:display 4:is_empty 5:exit   "))
  if x==1:
    y=input("Enter the element to added to the stack   ")
    S1.push(y)
  elif x==2:
    S1.pop()
  elif x==3:
    S1.disp()
  elif x==4:
    S1.is_empty()
  elif x==5:
    break;
  else:
    print("wrong choice")
```
